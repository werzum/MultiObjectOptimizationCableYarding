```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


sns.set(rc={'figure.figsize':(14,10)})
```

```python
with open("1_3.csv") as file:
        table_1_3 = pd.read_csv(file)
```

```python
table_1_3
```

```python
unique_country = table_1_3["Country"].value_counts(sort=False)
country_names = table_1_3["Unnamed: 6"].iloc[1:45]

fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(country_names, unique_country.values, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
# x = np.arange(len(country_names))  # the label locations
# fig, ax = plt.subplots()
# ax.bar(unique_country.values)
```
